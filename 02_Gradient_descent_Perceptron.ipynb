```python
from sklearn.datasets import make_classification
import numpy as np
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_clusters_per_class=1, n_redundant=0, random_state=41, n_classes=2, hypercube=False, class_sep=15)
```

```python
X[:5], y[:5]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
plt.show()
```

```python
def Perceptron(X, y):
  w1 = w2 = b = 1
  lr = 0.1
  epochs = 1000

  for n in range(epochs):
    for i in range(X.shape[0]):
      # finding predicted value
      z = w1*X[i][0] + w2*X[i][1] + b

      if z*y[i] < 0:
        w1 = w1 + lr*y[i]*X[i][0]
        w2 = w2 + lr*y[i]*X[i][1]
        b = b + lr*y[i]

  return w1, w2, b
```

```python
w1, w2, b = Perceptron(X, y)
m = -w1/w2
c = -b/w2
m, c
```

```python
x_input = np.linspace(-2,0,100)
y_output = m*x_input + c
max(x_input),min(x_input)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:, 1], c=y, cmap='winter', s=100)
plt.plot(x_input, y_output, color='red')
plt.show()
```
