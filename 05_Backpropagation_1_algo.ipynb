```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]], columns=['cgpa','profile_score','lpa'])
df
```

```python
def initialize_parameters(layer_dims):
  np.random.seed(3)
  parameters = {}
  L = len(layer_dims)

  for i in range(1,L):
    parameters['W'+str(i)] = np.ones((layer_dims[i-1],layer_dims[i]))*0.1
    parameters['b'+str(i)] = np.zeros((layer_dims[i],1))
  return parameters
```

```python
def linear_forward(A_prev, W, b):
  z = np.dot(W.T, A_prev) + b
  return z
```

```python
# forward propagation
def L_layer_forward(X, parameters):
  A = X
  L = len(parameters)//2   # number of layers in the neural networks

  for i in range(1, L+1):
    A_prev = A
    W = parameters['W'+str(i)]
    b = parameters['b'+str(i)]

    # print('A'+str(i-1)+':',A_prev)
    # print('W'+str(i)+':',W)
    # print('b'+str(i)+':',b)
    # print('-'*20)

    A = linear_forward(A_prev, W, b)
    # print('A'+str(i)+':',A)
    # print('*'*20)

  return A_prev, A
```

```python
# updating the weights using gradient descent

def update_parameters(parameters, y, y_hat, A_prev, X, lr):

  new_parameters = parameters

  new_parameters['W2'][0][0] = parameters['W2'][0][0]+lr*2*(y-y_hat)*A_prev[0][0]
  new_parameters['W2'][1][0] = parameters['W2'][1][0]+lr*2*(y-y_hat)*A_prev[1][0]
  new_parameters['b2'][0][0] = parameters['b2'][0][0]+lr*2*(y-y_hat)

  new_parameters['W1'][0][0] = parameters['W1'][0][0]+lr*2*(y-y_hat)*parameters['W2'][0][0]*X[0][0]
  new_parameters['W1'][0][1] = parameters['W1'][0][1]+lr*2*(y-y_hat)*parameters['W2'][0][0]*X[1][0]
  new_parameters['b1'][0][0] = parameters['b1'][0][0]+lr*2*(y-y_hat)*parameters['W2'][0][0]

  new_parameters['W1'][1][0] = parameters['W1'][1][0]+lr*2*(y-y_hat)*parameters['W2'][1][0]*X[0][0]
  new_parameters['W1'][1][1] = parameters['W1'][1][1]+lr*2*(y-y_hat)*parameters['W2'][1][0]*X[1][0]
  new_parameters['b1'][1][0] = parameters['b1'][1][0]+lr*2*(y-y_hat)*parameters['W2'][1][0]

  return new_parameters
```

```python
# training the model now
parameters = initialize_parameters([2,2,1])
parameters
epochs = 75
lr = 0.001

for i in range(epochs):
  avg_mse = 0
  for j in range(df.shape[0]):
    n = np.random.randint(0,df.shape[0])
    X = df[['cgpa','profile_score']].values[n].reshape(2,1)
    y = df[['lpa']].values[n][0]

    A, y_hat = L_layer_forward(X, parameters)
    parameters = update_parameters(parameters, y, y_hat[0][0], A, X, lr )

    avg_mse += (y - y_hat[0][0])**2
  print("epoch-",i+1,": loss :",avg_mse/df.shape[0])
  print('*'*40)
```

```python
# now using keras
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(2, activation='linear', input_dim=2))
model.add(Dense(1, activation='linear'))
```

```python
new_weights = [np.array([[0.1,0.1],[0.1,0.1]], dtype=np.float32),
               np.array([0,0], dtype=np.float32),
               np.array([[0.1],[0.1]], dtype=np.float32),
               np.array([0], dtype=np.float32)]
```

```python
model.set_weights(new_weights)
```

```python
model.get_weights()
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='mean_squared_error', optimizer=optimizer)
```
